```python
# Analisis de datos countries 
```

```python
data = pd.read_csv("countries.csv")
```

```python
data
```

```python
Ecuador = data[data.country == "Ecuador"]
```

```python
Ecuador
```

```python
Peru = data[data.country == "Peru"]
```

```python
Peru
```

```python
plt.plot(Ecuador.year, Ecuador.population / 10**6)
plt.plot(Peru.year, Peru.population /10**6)
plt.legend(["Ecuador", "Peru"])
plt.title("Poblacion Ecuador y Peru")
plt.xlabel("Ao")
plt.ylabel("Millones de aos")
plt.show()
```

```python
Ecuador.population / Ecuador.population.iloc[0]
```

```python
Ecuador.population / Ecuador.population.iloc[0]*100
```

```python
plt.plot(Ecuador.year, Ecuador.population / Ecuador.population.iloc[0]*100)
plt.plot(Peru.year, Peru.population / Peru.population.iloc[0]*100)
plt.legend(["Ecuador", "Peru"])
plt.title("Poblacion Ecuador y Peru")
plt.xlabel("Ao")
plt.ylabel("Crecimiento de la poblacion (1er ao 100)")
plt.show()
```
